```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = 0.3          # Probability of success
samples = 1000  # Number of observations
```

```python
data = np.random.geometric(p, samples)
```

```python
experimental_mean = np.mean(data)

print("Experimental Mean:", experimental_mean)
```

```python
# For k = 3
prob_x_gt_3 = np.sum(data > 3) / samples

# For k = 5
prob_x_gt_5 = np.sum(data > 5) / samples

print("P(X > 3) ≈", prob_x_gt_3)
print("P(X > 5) ≈", prob_x_gt_5)
```

```python
#let m and n be-
m = 2
n = 3
```

```python
# X > m+n
count_mn = np.sum(data > (m + n))

# X > m
count_m = np.sum(data > m)

left_side = count_mn / count_m
```

```python
right_side = np.sum(data > n) / samples
```

```python
print("Left Side P(X > m+n | X > m):", left_side)
print("Right Side P(X > n):", right_side)
```

```python
max_value = np.max(data)

x_values = np.arange(1, max_value + 1)

pmf = []

for x in x_values:
    prob = np.sum(data == x) / samples
    pmf.append(prob)
```

```python
plt.bar(x_values, pmf)

plt.xlabel("Number of Trials (X)")
plt.ylabel("Probability")
plt.title("Experimental PMF of Geometric Distribution")

plt.grid(True)
plt.show()
```
