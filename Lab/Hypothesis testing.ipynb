Question 2
A bottling company claims that the mean volume of juice bottles is 500 ml. The population
standard deviation is known to be 12 ml. A random sample of 64 bottles has a mean volume of
496 ml. At 5% significance level, test whether the company’s claim is true.
Implementation Tasks:
1. State 𝐻0and 𝐻1
2. Compute Z-statistic manually in Python
 3. Compute p-value using scipy.stats.norm
4. Interpret result

```python
#Q1
import numpy as np
from scipy import stats

data = [3.1, 3.4, 3.0, 3.3, 3.5, 3.2, 3.6, 3.1, 3.0, 3.3]

t_stat, p_value = stats.ttest_1samp(data, popmean=3.2)

print("t-statistic:", t_stat)
print("p-value:", p_value)


mean = np.mean(data)
std = np.std(data, ddof=1)  # sample std
n = len(data)

t_manual = (mean - 3.2) / (std / np.sqrt(n))

print("Manual t-statistic:", t_manual)

#Interpretation - If p-value < 0.05 → Reject H₀ else If p-value ≥ 0.05 → Fail to reject H₀


```

```python
#Q2)
import numpy as np
from scipy.stats import norm

x_bar = 496
mu = 500
sigma = 12
n = 64

z = (x_bar - mu) / (sigma / np.sqrt(n))

# two-tailed test
p_value = 2 * norm.cdf(-abs(z))

print("Z-statistic:", z)
print("p-value:", p_value)
```
