A Binomial random variable represents the number of successes in multiple independent
Bernoulli trials.
1. Simulate a Binomial random variable with:
o Number of trials 𝑛 = 10
o Probability of success 𝑝 = 0.5
2. Generate 1000 samples.
3. Estimate the experimental Probability Mass Function (PMF).
4. Compute and compare:
o Theoretical mean and variance
o Experimental mean and variance
5. Visualize the PMF using a bar chart.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10        # Number of trials
p = 0.5       # Probability of success
samples = 1000  # Number of samples
```

```python
data = np.random.binomial(n, p, samples)
```

```python
x_values = np.arange(0, n+1)
print("Possible values:", x_values)
```

```python
experimental_pmf = []

for x in x_values:
    probability = np.sum(data == x) / samples
    experimental_pmf.append(probability)
```

```python
theoretical_mean = n * p
theoretical_variance = n * p * (1 - p)
```

```python
experimental_mean = np.mean(data)
experimental_variance = np.var(data)
```

```python
print("Theoretical Mean:", theoretical_mean)
print("Experimental Mean:", experimental_mean)

print("Theoretical Variance:", theoretical_variance)
print("Experimental Variance:", experimental_variance)

print("\nExperimental PMF:")

for x, prob in zip(x_values, experimental_pmf):
    print(f"P(X = {x}) ≈ {prob:.4f}")
```

```python
plt.bar(x_values, experimental_pmf)

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Experimental PMF of Binomial Distribution")

plt.grid(True)
plt.show()
```
