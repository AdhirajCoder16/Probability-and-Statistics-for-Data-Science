1. A school wants to compare the effectiveness of two different teaching methods. Two
groups of students are randomly selected. The test scores of students taught using Method
A are: 72, 75, 78, 70, 74, 73. The test scores of students taught using Method B are: 68,
65, 70, 66, 69, 67. Assuming the samples are independent and drawn from normally
distributed populations, test whether there is a significant difference in the mean scores of
the two methods at the 5% significance level. Perform an independent two-sample t-testin Python and interpret the results.

2. A fitness trainer wants to evaluate the effectiveness of a 6-week weight loss program.

follows:
Before: 82, 85, 78, 90, 88, 76
After: 78, 82, 75, 85, 84, 73

 At the 5% level of significance, test whether the program significantly reduces weight.
Implement a paired t-test in Python, compute the test statistic and p-value, and state your

 3. A researcher wants to examine whether gender and product preference are independent.
A survey is conducted, and the following data are collected:
 • 40 males prefer Product A, 30 prefer Product B, and 20 prefer Product C.
• 35 females prefer Product A, 45 prefer Product B, and 30 prefer Product C.
Using a 5% significance level, test whether gender and product preference are independent.
Perform a Chi-square test of independence in Python, compute the Chi-square statistic,
degrees of freedom, and p-value, and interpret the results.

```python
from scipy import stats
import numpy as np

# Data
method_A = np.array([72, 75, 78, 70, 74, 73])
method_B = np.array([68, 65, 70, 66, 69, 67])

# Independent t-test
t_stat, p_value = stats.ttest_ind(method_A, method_B)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
from scipy import stats

# Data
before = np.array([82, 85, 78, 90, 88, 76])
after = np.array([78, 82, 75, 85, 84, 73])

# Paired t-test
t_stat, p_value = stats.ttest_rel(before, after)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
from scipy import stats

# Data
before = np.array([82, 85, 78, 90, 88, 76])
after = np.array([78, 82, 75, 85, 84, 73])

# Paired t-test
t_stat, p_value = stats.ttest_rel(before, after)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
from scipy.stats import chi2_contingency

# Contingency table
data = np.array([[40, 30, 20],
                 [35, 45, 30]])

chi2, p, dof, expected = chi2_contingency(data)

print("Chi-square:", chi2)
print("Degrees of freedom:", dof)
print("p-value:", p)
```
