Heights of students are normally distributed with
μ = 170 cm, σ = 8 cm.
1. What percentage are taller than 180 cm?
2. What percentage are between 160 and 175 cm?
3. Find height corresponding to top 5%.

```python
from scipy.stats import norm

# Given values
mu = 170
sigma = 8
```

```python
# Question 1: Taller than 180 cm
z1 = (180 - mu) / sigma
prob_taller_180 = 1 - norm.cdf(z1)

print("Percentage taller than 180 cm:", prob_taller_180 * 100, "%")
```

```python
# Question 2: Between 160 and 175 cm
z2 = (160 - mu) / sigma
z3 = (175 - mu) / sigma

prob_between = norm.cdf(z3) - norm.cdf(z2)

print("Percentage between 160 and 175 cm:", prob_between * 100, "%")
```

```python
# Question 3: Height for top 5%
z_95 = norm.ppf(0.95)

height_top_5 = mu + z_95 * sigma

print("Height corresponding to top 5%:", height_top_5, "cm")
```
