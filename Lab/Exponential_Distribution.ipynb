The time (in minutes) between arrivals of customers at a service desk follows an Exponential
distribution with an average arrival time of 4 minutes.
Tasks
1. Identify the rate parameter .
2. Using Python, compute:
Probability that the next customer arrives within 3 minutes.
Probability that the waiting time exceeds 8 minutes.
3. Calculate the theoretical mean and variance.
4. Simulate 10,000 inter-arrival times and calculate:
Empirical mean and Empirical variance
5. Verify the memoryless property using conditional probability.
6. Plot:
Histogram of simulated data
Theoretical PDF overlaid on the histogram

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean_time = 4
lam = 1 / mean_time

print("Rate parameter (lambda):", lam)
```

```python
prob_less_3 = 1 - np.exp(-lam * 3)
prob_greater_8 = np.exp(-lam * 8)

print("P(X < 3) =", prob_less_3)
print("P(X > 8) =", prob_greater_8)
```

```python
theoretical_mean = 1 / lam
theoretical_variance = 1 / (lam ** 2)

print("Theoretical Mean:", theoretical_mean)
print("Theoretical Variance:", theoretical_variance)
```

```python
samples = 10000
data = np.random.exponential(1 / lam, samples)
```

```python
empirical_mean = np.mean(data)
empirical_variance = np.var(data)

print("Empirical Mean:", empirical_mean)
print("Empirical Variance:", empirical_variance)
```

```python
#let m and n be
m = 3
n = 2
```

```python

count_mn = np.sum(data > (m + n))
count_m = np.sum(data > m)

left_side = count_mn / count_m
right_side = np.sum(data > n) / samples

print("P(X > m+n | X > m):", left_side)
print("P(X > n):", right_side)
```

```python
x = np.linspace(0, np.max(data), 1000)
pdf = lam * np.exp(-lam * x)

plt.hist(data, bins=40, density=True)
plt.plot(x, pdf)

plt.xlabel("Inter-arrival Time (Minutes)")
plt.ylabel("Density")
plt.title("Exponential Distribution: Histogram and PDF")

plt.grid(True)
plt.show()
```
