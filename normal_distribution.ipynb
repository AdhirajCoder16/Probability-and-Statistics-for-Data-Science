```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate Normal Distribution Data
mu = 100
sigma = 15
samples = 1000

data = np.random.normal(mu, sigma, samples)
```

```python
#Plot Histogram
plt.hist(data, bins=30, density=True)

plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram of Normal Distribution (μ=100, σ=15)")

plt.grid(True)
plt.show()
```

```python
#Compute Sample Mean and Standard Deviation
sample_mean = np.mean(data)
sample_std = np.std(data)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
```

```python
#Ranges
lower_1 = mu - sigma
upper_1 = mu + sigma

lower_2 = mu - 2*sigma
upper_2 = mu + 2*sigma

lower_3 = mu - 3*sigma
upper_3 = mu + 3*sigma
```

```python
#Calculate Percentages in Each Range
percent_1 = np.sum((data >= lower_1) & (data <= upper_1)) / samples * 100
percent_2 = np.sum((data >= lower_2) & (data <= upper_2)) / samples * 100
percent_3 = np.sum((data >= lower_3) & (data <= upper_3)) / samples * 100

print("Within μ ± 1σ:", percent_1, "%")
print("Within μ ± 2σ:", percent_2, "%")
print("Within μ ± 3σ:", percent_3, "%")
```
