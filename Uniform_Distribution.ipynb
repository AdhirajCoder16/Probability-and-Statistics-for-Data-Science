```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 10   # 10 AM
b = 12   # 12 PM
```

```python
# Probability before 11 AM
prob_before_11 = (11 - a) / (b - a)

# Probability between 10:30 and 11:30
prob_between = (11.5 - 10.5) / (b - a)

print("P(X < 11) =", prob_before_11)
print("P(10.5 < X < 11.5) =", prob_between)
```

```python
theoretical_mean = (a + b) / 2
theoretical_variance = (b - a)**2 / 12

print("Theoretical Mean:", theoretical_mean)
print("Theoretical Variance:", theoretical_variance)
```

```python
samples = 10000

data = np.random.uniform(a, b, samples)
```

```python
sample_mean = np.mean(data)
sample_variance = np.var(data)

print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
```

```python
plt.hist(data, bins=30, density=True)

plt.xlabel("Arrival Time (Hours)")
plt.ylabel("Density")
plt.title("Histogram of Uniform Distribution (10 AM to 12 PM)")

plt.grid(True)
plt.show()
```

```python
print("\nComparison:")

print("Theoretical Mean:", theoretical_mean)
print("Sample Mean:", sample_mean)

print("Theoretical Variance:", theoretical_variance)
print("Sample Variance:", sample_variance)
```
